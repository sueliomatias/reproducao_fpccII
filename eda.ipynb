```python
import pandas as pd
import numpy as np
from plotnine import *
import os
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

```python
# diretório onde estão os arquivos
directory = "dataset"

# inicializa um DataFrame vazio
combined_df = pd.DataFrame()

# itera sobre os arquivos do diretório
for filename in os.listdir(directory):
    if filename.endswith(".txt"):
        file_path = os.path.join(directory, filename)
        
        # lêr o arquivo e armazena em um DataFrame
        df = pd.read_csv(file_path, delimiter="\t")
        
        # concatena o DataFrame lido com o DataFrame principal
        combined_df = pd.concat([combined_df, df], ignore_index=True)

# salva o DataFrame principal em um arquivo CSV
combined_df.to_csv("artigos.csv", index=False)
combined_df.to_csv("artigos.txt", index=False)
```

### Figura 2 – Evolução da posição dos países de acordo com o número de publicações indexadas na WoS

```python
# Ajustar a expressão regular para capturar corretamente o país
df['country'] = df['RP'].str.extract(r',\s*([A-Za-z\s]+(?:USA|China|Brazil|France|Germany|India|Japan|Canada|Australia|Italy|UK|Peoples R China))\.?$')

# Agrupar por ano e país para contar o número de publicações
df_country_year = df.groupby(['PY', 'country']).size().unstack().fillna(0)

# Calcular a posição dos países ao longo do tempo
df_country_rank = df_country_year.rank(axis=1, method='min', ascending=False)

# Selecionar os 10 principais países com mais publicações em geral
top_countries = df_country_year.sum().sort_values(ascending=False).head(10).index

# Gerar gráfico
plt.figure(figsize=(12, 8))
for country in top_countries:
    plt.plot(df_country_rank.index, df_country_rank[country], label=country)
# plt.gca().invert_yaxis()
plt.title('Evolução da Posição dos Países de Acordo com o Número de Publicações Indexadas na WoS')
plt.xlabel('Ano')
plt.ylabel('Posição')
plt.legend()
plt.show()
```

### Figura 4 – Rede de coocorrência das 50 palavras-chave de maior ocorrência nas publicações indexadas na WoS

```python
import networkx as nx
from collections import Counter
from itertools import combinations

# Extrair palavras-chave
keywords = df['DE'].dropna().str.split('; ').explode()
keyword_counts = Counter(keywords)
top_keywords = [kw for kw, count in keyword_counts.most_common(50)]
# Filtrar as palavras-chave top 100 para criar a rede de coocorrência
valid_keywords = set(top_keywords)

# exibir as 100 palavras-chave mais frequentes
keyword_counts_table = pd.DataFrame.from_dict(keyword_counts, orient='index', columns=['Count'])
keyword_counts_table.index.name = 'Keyword'
display(keyword_counts_table.nlargest(100, 'Count'))

# Criar rede de coocorrência ajustada
cooccurrence = Counter()
for keywords_list in df['DE'].dropna().str.split('; '):
    filtered_keywords = [kw for kw in keywords_list if kw in valid_keywords]
    for pair in combinations(filtered_keywords, 2):
        cooccurrence[frozenset(pair)] += 1

G = nx.Graph()
for pair, count in cooccurrence.items():
    if len(pair) == 2:
        kw1, kw2 = pair
        G.add_edge(kw1, kw2, weight=count)

# Gerar rede de coocorrência
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G, k=0.3)
nx.draw_networkx(G, pos, with_labels=True, node_size=50, font_size=8, edge_color='#86989b3b')
plt.title('Rede de Coocorrência das 100 Palavras-chave de Maior Ocorrência nas Publicações Indexadas na WoS')
plt.show()
```

### Figura 7 – Rede de colaboração entre Instituições com maior número de publicações indexadas na WoS

```python
# Selecionar as principais instituições para uma visualização simplificada
top_institutions = df['C1'].dropna().str.split('; ').explode().value_counts().head(20).index
collaborations = df['C1'].dropna().str.split('; ').tolist()
collaborations = [list(set(insts).intersection(top_institutions)) for insts in collaborations if len(set(insts).intersection(top_institutions)) > 1]

# Criar rede de colaboração
G = nx.Graph()
for institutions in collaborations:
    for pair in combinations(institutions, 2):
        if G.has_edge(*pair):
            G[pair[0]][pair[1]]['weight'] += 1
        else:
            G.add_edge(*pair, weight=1)

# Gerar gráfico
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G, k=0.3)
edges = G.edges(data=True)
weights = [edge[2]['weight'] for edge in edges]
nx.draw_networkx(G, pos, with_labels=True, node_size=300, font_size=10, edge_color=weights, edge_cmap=plt.cm.Blues)
plt.title('Rede de Colaboração entre Instituições com Maior Número de Publicações Indexadas na WoS')
plt.show()
```

```python
import numpy as np
import re

# Termos relacionados à IA
ai_terms = ["artificial intelligence", "AI", "machine learning", "deep learning", "neural network"]

# Criar uma máscara booleana para selecionar as publicações relacionadas à IA
mask = df['TI'].str.contains('|'.join(ai_terms), case=False, na=False) | df['DE'].str.contains('|'.join(ai_terms), case=False, na=False)

# Filtrar os dados
ai_publications = df[mask]

# Extrair o país de cada publicação
# Assumiremos que o campo 'C1' (endereços) contém informações sobre os países

def extract_countries(addresses):
    countries = re.findall(r'\b[A-Z][A-Z]+\b', addresses)
    return countries

# Aplicar a extração de países
ai_publications['Countries'] = ai_publications['C1'].apply(lambda x: extract_countries(x) if pd.notnull(x) else [])

# Expandir a coluna de países
countries_expanded = ai_publications.explode('Countries')

# Contar o número de publicações por país
publications_by_country = countries_expanded['Countries'].value_counts()

publications_by_country
```
