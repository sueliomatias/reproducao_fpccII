```python
import pandas as pd
import numpy as np
from plotnine import *
import os
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

```python
# diretório onde estão os arquivos
directory = "dataset"

# inicializa um DataFrame vazio
combined_df = pd.DataFrame()

# itera sobre os arquivos do diretório
for filename in os.listdir(directory):
    if filename.endswith(".txt"):
        file_path = os.path.join(directory, filename)
        
        # lêr o arquivo e armazena em um DataFrame
        df = pd.read_csv(file_path, delimiter="\t")
        
        # concatena o DataFrame lido com o DataFrame principal
        combined_df = pd.concat([combined_df, df], ignore_index=True)

# salva o DataFrame principal em um arquivo CSV
combined_df.to_csv("artigos.csv", index=False)
combined_df.to_csv("artigos.txt", index=False)
```

### Figura 1 – Número de publicações indexadas na WoS, por país

```python
# Ajustar a expressão regular para capturar corretamente o país
df['country'] = df['C1'].str.extract(r',\s*([A-Z][A-Za-z\s]*[A-Za-z])\.?$')

# Substituir valores NaN na coluna 'country' com a expressão fornecida
df.loc[df['country'].isna(), 'country'] = df['C1'].apply(lambda x: ', USA' if isinstance(x, str) and 'USA' in x else x)
df['country'] = df['country'].str.replace(r'.*\sUSA$', 'USA', regex=True)

# Contagem de publicações por país
df_country_rp = df['country'].value_counts().head(25).sort_values(ascending=True)

# Gerar gráfico na horizontal
plt.figure(figsize=(12, 10))
ax = df_country_rp.plot(kind='barh')
plt.title('Figura 1 - Número de publicações sobre LLMs indexadas na WoS, por país')
plt.xlabel('Número de Publicações')
plt.ylabel('Países')

# adicionar rótulos com os valores
for p in ax.patches:
    ax.annotate(str(p.get_width()), (p.get_width(), p.get_y() + p.get_height() / 2), ha='left', va='center')

plt.show()
```

Os dados das publicações indexadas na Web of Science (WoS) destacam os Estados Unidos (1561) artigos e a China (528 artigos) como os principais contribuintes à produção científica global. Na Europa, Alemanha (252), Inglaterra (218), Itália (116) e França (68) se destacam, refletindo seus fortes investimentos em pesquisa e desenvolvimento (P&D). Destque também para o Canadá (137).

A Ásia também tem mostrado um crescimento significativo, com a Índia (143), Coreia do Sul (125), Japão (116) e Singapura (62) contribuindo de maneira notável. Isso indica uma expansão da produção científica na região.

A presença de publicações de países de todos os continentes, como Austrália (107) e Brasil (28), revela um engajamento global na produção científica.

### Figura 6 – Rede de coocorrência das 50 palavras-chave de maior ocorrência nas publicações indexadas na WoS

```python
import networkx as nx
from collections import Counter
from itertools import combinations

# Extrair palavras-chave
keywords = df['DE'].dropna().str.split('; ').explode()
keyword_counts = Counter(keywords)
top_keywords = [kw for kw, count in keyword_counts.most_common(50)]
# Filtrar as palavras-chave top 100 para criar a rede de coocorrência
valid_keywords = set(top_keywords)

# exibir as 100 palavras-chave mais frequentes
keyword_counts_table = pd.DataFrame.from_dict(keyword_counts, orient='index', columns=['Count'])
keyword_counts_table.index.name = 'Keyword'
display(keyword_counts_table.nlargest(100, 'Count'))

# Criar rede de coocorrência ajustada
cooccurrence = Counter()
for keywords_list in df['DE'].dropna().str.split('; '):
    filtered_keywords = [kw for kw in keywords_list if kw in valid_keywords]
    for pair in combinations(filtered_keywords, 2):
        cooccurrence[frozenset(pair)] += 1

G = nx.Graph()
for pair, count in cooccurrence.items():
    if len(pair) == 2:
        kw1, kw2 = pair
        G.add_edge(kw1, kw2, weight=count)

# Gerar rede de coocorrência
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G, k=0.3)
nx.draw_networkx(G, pos, with_labels=True, node_size=50, font_size=8, edge_color='#86989b3b')
plt.title('Rede de Coocorrência das 100 Palavras-chave de Maior Ocorrência nas Publicações Indexadas na WoS')
plt.show()
```
